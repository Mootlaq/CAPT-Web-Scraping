### Libraries and dependencies
- Pandas
- BeautifulSoup
- Selenium
    - **Note:** Selenium requires a driver to interface with the chosen browser. Firefox, for example, requires geckodriver, which needs to be installed before the below examples can be run.

```python
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
```

```python
def scrape_winning_tenders(webpage):
    from selenium.webdriver.firefox.options import Options #Used so that selenium doesn't open a Firefox window
    Options = Options()
    Options.headless = True #Yes, no Firefox window!
    driver = webdriver.Firefox(options=Options)
    driver.get(webpage)

    html = driver.page_source
    soup = BeautifulSoup(html)
    
    #Retrieve the table
    inner_table = soup.findAll('div', {"class":"custom-table"}) 
    
    #Retrieve table header and create a list of columns names
    head = soup.find('div', {"class":"table-row head"}) 
    column_name = head.find_all('div', {"class":"table-cell"}) #working
    columns = []
    for i in range(len(column_name)):
        columns.append(column_name[i].text)
        
    # Retrieve table records
    rows = soup.findAll('div', {"class":"table-row tbody"})
    records= []
    for row in rows:
        record = []
        cells = row.findAll('div', {"class":"table-cell"})
        for cell in cells:
            record.append(cell.text)
        records.append(record)
        
    df = pd.DataFrame(records, columns=columns) #Create Dataframe
    return df
```

```python
pages = [i for i in range(1,6)] # Create a list of the pages number. Here, it's five
original_url = "https://capt.gov.kw/ar/tenders/winning-bids/?page={}" #Winning bids page.
df = pd.DataFrame() # Create an empty Dataframe.
for page in pages:
    url = original_url.format(page) #add the page number to the url.
    df1 = scrape_winning_tenders(url) #scrape, baby.
    df = pd.concat([df,df1], ignore_index=True) #concat dataframes. 
```
